Vizualization for Univariate Analysis

```python
plt.figure(figsize=(18,10))
plt.subplot(2,2,1)
sns.histplot(x=df['price'], data=df, kde=True)
plt.title('Histogram Showing the Distribution of Price of Housing')
plt.subplot(2,2,2)
sns.boxplot(x=df['price'],data=df, color='red',orient='v')
plt.title('Boxplot Showing the Distribution of Price of Housing')
plt.subplot(2,2,3)
sns.histplot(x=df['area'], data=df,kde=True)
plt.title('Histogram Showing the Distribution of Price of Housing')
plt.subplot(2,2,4)
sns.boxplot(x=df['area'], data=df, color='yellow',vert=False)
plt.title('Boxplot Showing the Distribution of Price of Housing')
plt.show()
```

Vizualization for the Bivariate Analysis

```python
#Create a scatterplot to show the relationship between area and price of housing
sns.scatterplot(x=df['area'], y=df['price'], data=df, color= 'r', marker='*', s=200)
plt.title('Scatterplot Showing the Relationship between Area and Price of Housing')
plt.xlabel('Area')
plt.ylabel('Price')
plt.show()
```
